```python
import pandas as pd
import sqlite3
```

```python
file='healthcare_dataset.csv'
ds=pd.read_csv(file)
```

```python
conn=sqlite3.connect('Health_Care')
ds.to_sql('table',conn,index=False,if_exists='replace')
```

```python
query= '''SELECT *
FROM "table"'''
result=pd.read_sql_query(query,conn)
result
```

```python
query= '''SELECT COUNT(*)
FROM "table"
WHERE Name= NULL '''
result=pd.read_sql_query(query,conn)
result

```

```python
query= '''SELECT COUNT(*)
FROM "table"
WHERE Name IS NOT NULL '''
result=pd.read_sql_query(query,conn)
result
```

```python
query = """  
SELECT   
    COUNT(*) - COUNT("Name") AS Missing_Name,  
    COUNT(*) - COUNT("Age") AS Missing_Age,  
    COUNT(*) - COUNT("Gender") AS Missing_Gender,  
    COUNT(*) - COUNT("Blood Type") AS Missing_Blood_Type,  
    COUNT(*) - COUNT("Medical Condition") AS Missing_Medical_Condition,  
    COUNT(*) - COUNT("Date Of Admission") AS Missing_Date_of_Admission,  
    COUNT(*) - COUNT("Doctor") AS Missing_Doctor,  
    COUNT(*) - COUNT("Hospital") AS Missing_Hospital,  
    COUNT(*) - COUNT("Insurance Provider") AS Missing_Insurance_Provider,  
    COUNT(*) - COUNT("Billing Amount") AS Missing_Billing_Amount,  
    COUNT(*) - COUNT("Room Number") AS Missing_Room_Number,  
    COUNT(*) - COUNT("Admission Type") AS Missing_Admission_Type,  
    COUNT(*) - COUNT("Discharge Date") AS Missing_Discharge_Date,  
    COUNT(*) - COUNT("Medication") AS Missing_Medication,  
    COUNT(*) - COUNT("Test Results") AS Missing_Test_Results  
FROM "table"  
WHERE "Age" = 43 
"""
result=pd.read_sql_query(query,conn)
result
```

```python
query= '''SELECT COUNT(*)
FROM "table"
WHERE "Medical Condition" ="Cancer" '''
result=pd.read_sql_query(query,conn)
result
```

```python
query = '''
SELECT DISTINCT(Name)
from "table"
where "Medical Condition"="Cancer"
'''
result = pd.read_sql_query(query,conn)
result
```

```python
query = '''select * from "table"
where "Medical Condition" = "Cancer" and Age =43'''
result = pd.read_sql_query(query,conn)
result
```

```python
query = '''select * from "table"
where "Medical Condition" = "Cancer" and Age between 40 and 60'''
result = pd.read_sql_query(query,conn)
result
```

```python
query='''select Name , avg(Age)
from "table" 
where Name Like "%N%" AND "Medical Condition" in("Cancer","Asthma","Diabetes")
GROUP BY Name'''
result = pd.read_sql_query(query,conn)
result
```

```python
query='''select Name , Age,CASE
WHEN Age<50 THEN "Young"
WHEN Age>50 THEN "Old"
END age_status
from "table" 
where Name LIKE "%N%"'''
result = pd.read_sql_query(query,conn)
result
```

```python
query='''select *,AVG(Age) OVER(partition by "Medical Condition")
from "table"
'''
result = pd.read_sql_query(query,conn)
result
```
